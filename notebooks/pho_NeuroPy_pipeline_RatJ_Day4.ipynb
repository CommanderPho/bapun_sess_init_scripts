## Path B: Klusta kwik → merged Phy export

Legacy per-shank Klusta/Phy curation lives in `klusta_day4_dir/Shank*/RatJDay4sh*.kwik`.

- **Good units**: `cluster_group == 3` (Phy curation labels). Do **not** use `si.read_klusta(..., exclude_cluster_groups=...)` — SpikeInterface's default kwik group mapping differs.
- **Shank8** is skipped until clustering completes (no clusters in kwik).
- **Compute cost**: SortingAnalyzer + `export_to_phy` on ~7 hr / ~1300 units is heavy. Use `PHY_EXPORT_OVERWRITE = "if_missing"` to avoid accidental re-runs.
- **Output**: `{basedir}/SORTING/klusta_kwik_merged_phy/` for `PhyIO` / NeuroPy downstream.

```python
import h5py
import spikeinterface.extractors as se
import spikeinterface.full as si
from spikeinterface.core import aggregate_units, NumpySorting
from spikeinterface.exporters import export_to_phy
from probeinterface.io import read_prb
from spikeinterface_pipeline.phy_export import bootstrap_cluster_info, write_bapun_params_py

RUN_KWIK_PHY_EXPORT = True
PHY_EXPORT_OVERWRITE = "if_missing"  # "if_missing" | "always"
N_JOBS = 8
DEFAULT_GAIN_TO_UV = 0.19499999284744263

klusta_phy_folder = (basedir / "SORTING" / "klusta_kwik_merged_phy").resolve()
klusta_analyzer_folder = (basedir / "SORTING" / "klusta_kwik_merged_sorting_analyzer").resolve()
dat_path = (basedir / f"{basename}.dat").resolve()
prb_path = (basedir / f"{basename}.prb").resolve()
job_kwargs = dict(n_jobs=N_JOBS, progress_bar=True, chunk_duration="4s")

```

```python

def _kwik_channel_group_key(kwik_file: h5py.File) -> str:
    channel_group_keys = sorted(kwik_file["channel_groups"].keys(), key=int)
    if len(channel_group_keys) != 1:
        raise ValueError(f"expected one channel group in {kwik_file.filename}, found {channel_group_keys}")
    return channel_group_keys[0]


def _load_good_units_from_kwik(kwik_path: Path, shank_id: int, sampling_frequency: float):
    with h5py.File(kwik_path, "r") as kwik_file:
        channel_group_key = _kwik_channel_group_key(kwik_file)
        clusters_group = kwik_file[f"channel_groups/{channel_group_key}"].get("clusters")
        if clusters_group is None or clusters_group.get("main") is None:
            return None
        clusters_main = clusters_group["main"]
        if len(clusters_main.keys()) == 0:
            return None
        spike_samples = kwik_file[f"channel_groups/{channel_group_key}/spikes/time_samples"][()]
        spike_clusters = kwik_file[f"channel_groups/{channel_group_key}/spikes/clusters/main"][()]
        good_cluster_ids = [int(cid) for cid in clusters_main.keys() if int(clusters_main[cid].attrs.get("cluster_group", -1)) == 3]
        if not good_cluster_ids:
            return None
        units_dict = {cluster_id: spike_samples[spike_clusters == cluster_id] for cluster_id in good_cluster_ids}
    sorting = NumpySorting.from_unit_dict(units_dict, sampling_frequency)
    sorting.set_property("shank", np.full(sorting.get_num_units(), shank_id, dtype=np.int64))
    sorting.set_property("local_cluster", np.array([int(u) for u in sorting.unit_ids], dtype=np.int64))
    return sorting


def _phy_export_exists(phy_folder: Path) -> bool:
    return phy_folder.is_dir() and (phy_folder / "params.py").is_file() and (phy_folder / "spike_clusters.npy").is_file() and (phy_folder / "cluster_info.tsv").is_file()


if not RUN_KWIK_PHY_EXPORT:
    print(f"RUN_KWIK_PHY_EXPORT=False; skipping kwik Phy export. Expected folder: {klusta_phy_folder}")
else:
    assert dat_path.is_file(), f"missing dat: {dat_path}"
    assert prb_path.is_file(), f"missing prb: {prb_path}"
    sortings, units_per_shank = [], {}
    for shank_id in range(1, 9):
        shank_dir = klusta_day4_dir / f"Shank{shank_id}"
        kwik_paths = sorted(shank_dir.glob("*.kwik")) if shank_dir.is_dir() else []
        if not kwik_paths:
            print(f"Shank{shank_id}: missing kwik — skipped")
            continue
        shank_sorting = _load_good_units_from_kwik(kwik_paths[0], shank_id=shank_id, sampling_frequency=float(dat_file_sampling_rate))
        if shank_sorting is None:
            print(f"Shank{shank_id}: no good clusters (cluster_group==3) — skipped")
            continue
        sortings.append(shank_sorting)
        units_per_shank[shank_id] = shank_sorting.get_num_units()
        print(f"Shank{shank_id}: loaded {shank_sorting.get_num_units()} good units from {kwik_paths[0].name}")
    if not sortings:
        raise RuntimeError("No kwik sortings loaded; cannot build merged Phy export.")
    combined = aggregate_units(sortings)
    print(f"Aggregated {combined.get_num_units()} units across {len(sortings)} shanks: {units_per_shank}")
    recording = se.read_binary(file_paths=dat_path.as_posix(), sampling_frequency=float(dat_file_sampling_rate), num_channels=n_channels, dtype="int16")
    recording.set_channel_gains(DEFAULT_GAIN_TO_UV)
    recording.set_channel_offsets(0)
    recording = recording.set_probe(read_prb(prb_path).probes[0], in_place=False)
    recording_filtered = si.bandpass_filter(recording)
    combined.register_recording(recording_filtered)
    run_analyzer = PHY_EXPORT_OVERWRITE == "always" or not klusta_analyzer_folder.is_dir()
    if run_analyzer:
        print(f"Creating SortingAnalyzer at {klusta_analyzer_folder} ...")
        sorting_analyzer = si.create_sorting_analyzer(sorting=combined, recording=recording_filtered, format="binary_folder", folder=klusta_analyzer_folder.as_posix(), overwrite=True)
        sorting_analyzer.compute(["random_spikes", "waveforms", "templates", "unit_locations"], **job_kwargs)
        if hasattr(sorting_analyzer, "save"):
            sorting_analyzer.save()
    else:
        print(f"Loading existing SortingAnalyzer from {klusta_analyzer_folder}")
        sorting_analyzer = si.load_sorting_analyzer(klusta_analyzer_folder)
    run_phy_export = PHY_EXPORT_OVERWRITE == "always" or not _phy_export_exists(klusta_phy_folder)
    if run_phy_export:
        print(f"Exporting Phy folder to {klusta_phy_folder} ...")
        klusta_phy_folder.mkdir(parents=True, exist_ok=True)
        sparsity = si.compute_sparsity(sorting_analyzer)
        export_to_phy(sorting_analyzer=sorting_analyzer, output_folder=klusta_phy_folder, copy_binary=False, remove_if_exists=True, sparsity=sparsity, verbose=True, **job_kwargs)
    else:
        print(f"Skipping export_to_phy; existing Phy export found at {klusta_phy_folder}")
    write_bapun_params_py(klusta_phy_folder, dat_path, n_channels=n_channels, sampling_rate=float(dat_file_sampling_rate))
    cluster_info = bootstrap_cluster_info(klusta_phy_folder, sorting_analyzer)
    cluster_info["group"] = "good"
    cluster_info.to_csv(klusta_phy_folder / "cluster_info.tsv", sep="\t", index=False)
    good_units_df = pd.DataFrame({"shank": combined.get_property("shank"), "local_cluster": combined.get_property("local_cluster")})
    good_units_df.to_csv(klusta_phy_folder / "good_units.tsv", sep="\t", index=False)
    print(f"Done. Phy folder: {klusta_phy_folder}")
    print(f"  units: {combined.get_num_units()}, spikes: {combined.to_spike_vector().size}, good_units.tsv rows: {len(good_units_df)}")


"""
Run:
phy template-gui  /media/halechr/BETAMAX1/Data/Bapun/RatJ/Day4Openfield/SORTING/klusta_kwik_merged_phy/params.py
Done. Phy folder: /media/halechr/BETAMAX1/Data/Bapun/RatJ/Day4Openfield/SORTING/klusta_kwik_merged_phy
  units: 1304, spikes: 18836895, good_units.tsv rows: 1304
"""
56m 15s
```

### Add in a buffer before and after each epoch if desired.

```python
def add_epoch_buffer(epoch_df: pd.DataFrame, buffer_sec: float or int or tuple or list):
    """Extend each epoch by buffer_sec before/after start/stop of each epoch"""
    if type(buffer_sec) in [int, float]:
        buffer_sec = (buffer_sec, buffer_sec)
    else:
        assert len(buffer_sec) == 2
        
    epoch_df['start'] -= buffer_sec[0]
    epoch_df['stop'] += buffer_sec[1]
```
